```python
import os
import random
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
```

```python
# Load your YOLO model
my_new_model = YOLO(r"D:\HaxS\YoloV8\data\results\200_epochs-3\weights\best.pt")
```

```python
# Specify the directory containing images
image_dir = r"D:\HaxS\YoloV8\data\test\images"
```

```python
# Get a list of all image files in the directory
image_files = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.jpeg', '.png'))]
```

```python
# Select a random image from the list
random_image_file = random.choice(image_files)
random_image_path = os.path.join(image_dir, random_image_file)
```

```python
# Read the selected image using cv2
new_image = cv2.imread(random_image_path)
```

```python
# Convert the image from BGR (OpenCV format) to RGB (matplotlib format)
new_image_rgb = cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB)
```

```python
# Perform prediction
new_results = my_new_model.predict(random_image_path, conf=0.2)  # Adjust confidence threshold
```

```python
# Plot the results
new_result_array = new_results[0].plot()
plt.figure(figsize=(12, 12))
plt.imshow(new_result_array)
plt.axis('off')  # Hide axis
plt.title(f"Results for: {random_image_file}")  # Title with the image name
plt.show()
```
